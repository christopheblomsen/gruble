Målet er da å lage en klasse, som tar i mot partikkel masser og hvor mange partikler det er som atributer.

Så skal vi ha minimum 2 metoder. Én plotter det vi har, og én som bruker [Euler-Cromer](https://en.wikipedia.org/wiki/Semi-implicit_Euler_method) på å regne ut utviklingen til posisjoner. Euler-Cromer er vanlig ODE som mange av dere kommer til å bli vell kjent med i senere kurs og den er på formen

\begin{align}
    v_{i+1} &= v_i + f(t_n, r_n)\cdot \Delta t\\
    r_{i+1} &= r_i + v_{i+1}\cdot\Delta t
\end{align}

Hvor $v$ er fartsvektor, $r$ er posisjonsvektor, $\Delta t$ er tidsteget og $f(t,r)$ er en funksjon. Denne funksjonen kommer til vårt tilfelle å være akselerasjonen som dere kan finne ett utrykk for med hjelp av Newtons andre lov.
